```python
##Pandas, Numpy, Matplotlib
import pandas as ps
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = ps.read_csv('voice.csv')
#selected_columns = data1[0:len(data1)-1]
data = data1.iloc[:,0:data1.shape[1]-1]
print(data.shape)
```

```python
data.head()
```

```python
correlation = data.corr()
plt.figure(figsize=(8,8))
sns.heatmap(correlation, vmax=1, square=True,annot=False,cmap='coolwarm')
```

```python
##Matrice centre
data_centre = data.sub(data.mean())
print(data_centre)
```

```python
##Matrice centre rduite
data_centre_reduite = data_centre.div(data_centre.std())
print(data_centre_reduite)
```

```python
##Matrice de covariance
matrice_cov=1/data.shape[0]*np.dot(data_centre_reduite.T,data_centre_reduite)
print(matrice_cov)
```

```python
##Valeurs et vecteurs propres
from numpy.linalg import eig
[Lambda,A]=eig(matrice_cov)
print("\n Valeurs propres ")
print(Lambda)
print("\n\n Vecteurs propres ")
#print(A)
```

```python
##Classement des valeurs propres
vp = np.sort(Lambda)[::-1]
print(vp)
```

```python
#Somme des valeurs propres
print(sum(vp))
```

```python
def pour_inertie(Lambda):
    L=[]
    for i in range(Lambda.shape[0]):
      s=Lambda[i]/sum(Lambda)*100
      L.append(s)
    return np.array(L)

# Pourcentage d'inertieA
pi = pour_inertie(vp)
print(pi)
```

```python
def pour_inertie_cumuler(x):
    L=[]
    s=0
    for i in range(x.shape[0]):
      s+=x[i]
      L.append(s)
    return np.array(L)

pic = pour_inertie_cumuler(pi)
print(pic)
```

# Ce code effectue une classification sur un ensemble de donnes en utilisant l'algorithme XGBoost.

### Importation des bibliothques :

pandas pour la manipulation des donnes, sklearn pour le prtraitement et les mtriques d'valuation, et xgboost pour le classificateur XGBoost.

```python
import pandas as pd
voice=pd.read_csv('voice.csv')
voice.head()
```

# fournit des informations statistiques sur l'ensemble de donnes.

```python
voice.describe()
```

Le code importe le module preprocessing de sklearn pour effectuer des tches de prtraitement des donnes.
# LabelEncoder()
utilis pour encoder les tiquettes de la colonne "label" du dataframe voice en valeurs numriques.
# le.classes_
affiche les classes distinctes dans les tiquettes encodes.


```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
voice["label"] = le.fit_transform(voice["label"])
le.classes_
```

# MinMaxScaler()
appliqu pour mettre  l'chelle les caractristiques numriques du dataframe voice dans une plage entre 0 et 1.

```python
voice[:]=preprocessing.MinMaxScaler().fit_transform(voice)
voice.head()
```

# Sparation des donnes d'entranement et de test :

```python
from sklearn.model_selection import train_test_split
import xgboost
```

# train_test_split()
utilise pour diviser l'ensemble de donnes prtrait en ensembles d'entranement et de test, avec une taille de test de 0,3 (30 % des donnes sont utilises pour les tests).

```python
train, test = train_test_split(voice, test_size=0.3)
```

```python
train.head()
```

# classify()
prend un modle, des donnes d'entranement et des donnes de test en entre.

Le modle est entran  l'aide de la mthode fit() sur les donnes d'entranement.


# classification_report()

utilis pour gnrer un rapport contenant la prcision, le rappel, le score F1 et le support pour chaque classe.

La fonction affiche le rapport de classification.

```python
def classify(model,x_train,y_train,x_test,y_test):
    from sklearn.metrics import classification_report
    target_names = ['female', 'male']
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print(classification_report(y_test, y_pred, target_names=target_names, digits=4))
```

les variables x_train, y_train, x_test et y_test pour stock les caractristiques d'entre et les tiquettes de sortie  la fois pour les donnes d'entranement et de test.

```python
x_train = train.iloc[:, :-1]
y_train = train["label"]
x_test = test.iloc[:, :-1]
y_test = test["label"]

model = xgboost.XGBClassifier()
classify(model,x_train,y_train,x_test,y_test)
```

```python
x_train2 = train[["meanfun","sd","median","Q25","Q75"]]
y_train2 = train["label"]
x_test2 = test[["meanfun","sd","median","Q25","Q75"]]
y_test2 = test["label"]

x_train2.head()
```

```python
model = xgboost.XGBClassifier()
classify(model,x_train2,y_train2,x_test2,y_test2)
```
